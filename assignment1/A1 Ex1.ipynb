## Question 1.2 - Cluster battery_power and ram

```python
from sklearn.preprocessing import MinMaxScaler

data = df[["battery_power", "ram"]]

# normalize the data to ensure that all dimensions are treated equally
# all columns should have the same impact on distance
mms = MinMaxScaler()
data_scaled = pd.DataFrame(mms.fit_transform(data), columns=["battery_power", "ram"])
data_scaled.describe()
```

```python
from sklearn.cluster import KMeans

k = 4
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(data_scaled)
# print the lables of the datapoints (which group they were assigned to)
print("Labels of the datapoints:", "\n", kmeans.labels_)
print("\n")
# print the coordinates of the centroids
print("Coorindates of the centroids:", "\n", kmeans.cluster_centers_)
```

```python
sns.scatterplot(x=data_scaled["battery_power"], y=data_scaled["ram"], hue=kmeans.labels_, palette="flare")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# silhouette_score indicates how far the datapoints are from the centroid on average
# mean of the silhouette coefficient for each datapoint (between +1 and -1)
# +1: instance is well inside own cluster and far from other clusters
# 0: instance is close to boundary
# -1: instance may have been assigned to wrong cluster
silhouette_score(data_scaled, kmeans.labels_)
```

```python
# Finding the optimal number of clusters
cluster_list = []
silhouette_list = []
# range from 1 to 5 as 1 cluster does not make sense
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=1) # kmeans with i clusters
    y_pred = kmeans.fit_predict(data_scaled) # applying k-means on the data

    cluster_list.append(i)
    silhouette_list.append(silhouette_score(data_scaled, kmeans.labels_))

print(silhouette_list)
score = pd.DataFrame({"Clusters": cluster_list, "Silhouette Score": silhouette_list})
print("silhouette_score for sepcified number of clusters:" "\n", score)

print("\n")
print("Maximum Silhouette Score and optimal number of clusters")
print(score[score["Silhouette Score"] == max(score["Silhouette Score"])])

sns.lineplot(x=score["Clusters"], y=score["Silhouette Score"])
plt.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

for i in range(2, 6):
    k = i+1
    kmeans = KMeans(n_clusters=k, random_state=1)
    y_pred = kmeans.fit_predict(data_scaled)

    visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')
    visualizer.fit(data_scaled)
    visualizer.show()

# silhouette diagram support the hypothesis that 4 clusters are ideal
# vertical dashed lines represent the silhouette score for each number of clusters
# when most of the instances in a cluster have a lower coefficient than this score (dashed line) then the cluster is rather bad -> instances are too close to other clusters
# while true for all illsutrations, esp. for k=4 most instances extend well beyond the dashed line -> good clusters
```

```python
from sklearn.cluster import DBSCAN

eps = 0.034
min_s = 5

dbscan = DBSCAN(eps=eps, min_samples=min_s)
y_pred = dbscan.fit_predict(data_scaled)

sns.scatterplot(data=data_scaled, x="battery_power", y="ram", hue=y_pred, palette="flare")
plt.title("min_sample="+str(min_s)+ " epsilon_range="+str(eps)+" clusters="+str(len(set(y_pred))-1))
plt.legend(facecolor='gray', frameon=True)
plt.show()

# pointless as whole chart is dense...
# finding clusters based on dense regions does not make sense
# even when specifying the hyperparameters, no meaningful clusters can be identfied
```
