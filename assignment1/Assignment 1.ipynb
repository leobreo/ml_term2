## Question 1.2 - battery_power and ram

```python
from sklearn.preprocessing import MinMaxScaler

data = df[["battery_power", "ram"]]

# normalize the data to ensure that all dimensions are treated equally
# all columns should have the same impact on distance
mms = MinMaxScaler()
data_scaled = pd.DataFrame(mms.fit_transform(data), columns=["battery_power", "ram"])
data_scaled.describe()
```

```python
from sklearn.cluster import KMeans

k = 4
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(data_scaled)
# print the lables of the datapoints (which group they were assigned to)
print("Labels of the datapoints:", "\n", kmeans.labels_)
print("\n")
# print the coordinates of the centroids
print("Coorindates of the centroids:", "\n", kmeans.cluster_centers_)
```

```python
sns.scatterplot(data_scaled, x="battery_power", y="ram", hue=kmeans.labels_, palette="flare")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# silhouette_score indicates how far the datapoints are from the centroid on average
# mean of the silhouette coefficient for each datapoint (between +1 and -1)
# +1: instance is well inside own cluster and far from other clusters
# 0: instance is close to boundary
# -1: instance may have been assigned to wrong cluster
silhouette_score(data_scaled, kmeans.labels_)
```

```python
# Finding the optimal number of clusters
cluster_list = []
silhouette_list = []
# range from 1 to 5 as 1 cluster does not make sense
for i in range(1, 10):
    k = i+1
    kmeans = KMeans(n_clusters=k, random_state=1)
    y_pred = kmeans.fit_predict(data_scaled)

    cluster_list.append(k)
    silhouette_list.append(silhouette_score(data_scaled, kmeans.labels_))

print(silhouette_list)
score = pd.DataFrame({"Clusters": cluster_list, "Silhouette Score": silhouette_list})
print("silhouette_score for sepcified number of clusters:" "\n", score)

print("\n")
print("Maximum Silhouette Score and optimal number of clusters")
print(score[score["Silhouette Score"] == max(score["Silhouette Score"])])

sns.lineplot(score, x="Clusters", y="Silhouette Score")
plt.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

for i in range(2, 6):
    k = i+1
    kmeans = KMeans(n_clusters=k, random_state=1)
    y_pred = kmeans.fit_predict(data_scaled)

    visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')
    visualizer.fit(data_scaled)
    visualizer.show()
```

```python
from sklearn.cluster import DBSCAN

eps = 0.034
min_s = 5

dbscan = DBSCAN(eps=eps, min_samples=min_s)
y_pred = dbscan.fit_predict(data_scaled)

sns.scatterplot(data=data_scaled, x="battery_power", y="ram", hue=y_pred, palette="flare")
plt.title("min_sample="+str(min_s)+ " epsilon_range="+str(eps)+" clusters="+str(len(set(y_pred))-1))
plt.legend(facecolor='gray', frameon=True)
plt.show()
```

Conducting the PCA
- X = UΣV(T)
- XND = XW(N)
- W(d) is the first d columns of V (first d principal components)

```python
# after PCA, the dimensions should explain 99% of the variance
exp_var = 0.99
pca = PCA(n_components=exp_var)

# X projected onto n dimensions to explain 99% of variance
X_com = pca.fit_transform(faces.data)

# unit vectors that define the first principal components:
print("# of unit verctors in each PCs:", "\n", pca.components_.T.shape[0])
# <unit vectors for PC1<
print("# of principal components:", "\n", pca.components_.T.shape[1])

print("\n")

# Reduced the number of features from 4096 to 260
print("400 images / datapoints and 260 principal components", "\n", X_com.shape)
pca.components_
```

Comparing reconstruction - manually and using sklearn

```python
import numpy as np
import sklearn.datasets, sklearn.decomposition

X = sklearn.datasets.load_iris().data
mu = np.mean(X, axis=0)

print(X[0,])
print(X.shape)

pca_ab = sklearn.decomposition.PCA(4)
pca_ab.fit(X)

# manual reconstruction
Xhat = np.dot(pca_ab.transform(X), pca_ab.components_) # [:,:nComp] [:nComp,:]
Xhat += mu

print(Xhat[0,])
print(pca_ab.components_.T)
print(pca_ab.explained_variance_ratio_.sum())

# reconstruction using sklearn
pca_ba = sklearn.decomposition.PCA(2)
x_trans = pca_ba.fit_transform(X) # reduced x
other_x = pca_ba.inverse_transform(x_trans) # reconstructed x

print(other_x[0,])
print(pca_ba.components_.T)
print(pca_ba.explained_variance_ratio_.sum())
```
